```python
# ============================================
# CONVERTIR TIMESTAMP A FECHA/HORA
# ============================================
```

```python
# La columna 'timecreated' contiene números que son timestamps Unix (segundos desde 1970)
# Los convertimos a formato de fecha y hora que podamos leer

# Convierte 'timecreated' de timestamp Unix a datetime
# unit='s' significa que los números están en segundos
df['datetime'] = pd.to_datetime(df['timecreated'], unit='s')

# Extrae solo la fecha (sin hora) de la columna datetime
df['date'] = df['datetime'].dt.date

# Extrae solo la hora (sin fecha) de la columna datetime
df['time'] = df['datetime'].dt.time

# Extrae el día de la semana (Monday, Tuesday, Wednesday, etc)
df['day_of_week'] = df['datetime'].dt.day_name()

# Extrae el nombre del mes (January, February, etc)
df['month'] = df['datetime'].dt.month_name()

# Muestra las nuevas columnas que creamos
print("NUEVAS COLUMNAS CREADAS:")
print(df[['timecreated', 'datetime', 'date', 'time', 'day_of_week', 'month']].head(10))
```

```python
# ============================================
# LIMPIAR VALORES NULOS
# ============================================
```

```python
# Vemos cuántos registros tenemos antes de limpiar
print(f"\nREGISTROS ANTES DE LIMPIAR: {len(df)}")

# Eliminamos las filas que tienen valores nulos en la columna 'department'
# dropna() elimina filas con valores nulos
# subset=['department'] especifica que solo queremos eliminar si 'department' está nulo
# inplace=True modifica el dataframe original (no crea una copia)
df.dropna(subset=['department'], inplace=True)

# Vemos cuántos registros quedaron después de limpiar
print(f"REGISTROS DESPUÉS DE LIMPIAR: {len(df)}")

# Calculamos cuántos registros eliminamos
print(f"REGISTROS ELIMINADOS: {22529 - len(df)}")
```

```python
# ============================================
# CONFIGURACIÓN INICIAL DE LOS GRÁFICOS
# ============================================
```

```python
# Establece el estilo de los gráficos
sns.set_style("whitegrid")
# Cambia la paleta de colores a algo más atractivo
sns.set_palette("husl")
```

```python
# ============================================
# GRÁFICO 1: TOP 15 DEPARTAMENTOS
# ============================================
```

```python
print("Creando gráfico 1: Top 15 Departamentos...")

# Prepara los datos: agrupa por departamento y cuenta ingresos
top_deptos = (df.groupby('department')
              .size()
              .reset_index(name='Ingresos')
              .sort_values(by='Ingresos', ascending=True)
              .tail(15))

# Crea la figura (gráfico)
# figsize=(12, 8) define el tamaño: 12 ancho, 8 alto
plt.figure(figsize=(12, 8))

# Crea un gráfico de barras horizontal
# Los nombres de departamentos van en el eje Y
# Los valores (ingresos) van en el eje X
plt.barh(top_deptos['department'], top_deptos['Ingresos'], color='steelblue', edgecolor='black')

# Etiquetas y título
plt.xlabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.ylabel('Departamento', fontsize=12, fontweight='bold')
plt.title('Top 15 Departamentos con Más Ingresos a la Plataforma', fontsize=14, fontweight='bold')

# Añade los valores encima de cada barra
for i, (depto, value) in enumerate(zip(top_deptos['department'], top_deptos['Ingresos'])):
    plt.text(value, i, f'  {value}', va='center', fontweight='bold')

# Ajusta el layout para que no se corte nada
plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 2: TOP 20 DOCENTES
# ============================================
```

```python
print("\nCreando gráfico 2: Top 20 Docentes...")

# Prepara los datos
top_docentes = (df.groupby('userid')
                .size()
                .reset_index(name='Ingresos')
                .sort_values(by='Ingresos', ascending=True)
                .tail(20))

# Crea la figura
plt.figure(figsize=(12, 8))

# Gráfico de barras horizontal
plt.barh(range(len(top_docentes)), top_docentes['Ingresos'], color='coral', edgecolor='black')

# Etiquetas y título
plt.xlabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.ylabel('Docente (ID)', fontsize=12, fontweight='bold')
plt.title('Top 20 Docentes con Más Ingresos a la Plataforma', fontsize=14, fontweight='bold')

# En lugar de mostrar los IDs largos, muestra números
# Los IDs van en la posición y el nombre es "Docente X"
plt.yticks(range(len(top_docentes)), 
           [f"Docente {i+1}" for i in range(len(top_docentes))])

# Añade los valores encima de cada barra
for i, value in enumerate(top_docentes['Ingresos']):
    plt.text(value, i, f'  {value}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 3: INGRESOS POR DÍA DE LA SEMANA
# ============================================
```

```python
print("\nCreando gráfico 3: Ingresos por Día de la Semana...")

# Prepara los datos en orden correcto
dias_orden = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dias_español = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

ingresos_dia = df['day_of_week'].value_counts().reindex(dias_orden)

# Crea la figura
plt.figure(figsize=(12, 6))

# Gráfico de barras vertical
# Usa una paleta de colores: colores más claros para fines de semana
colores = ['#1f77b4' if dia not in ['Saturday', 'Sunday'] else '#ff7f0e' 
           for dia in dias_orden]

bars = plt.bar(dias_español, ingresos_dia.values, color=colores, edgecolor='black', linewidth=2)

# Etiquetas y título
plt.xlabel('Día de la Semana', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.title('Ingresos a la Plataforma por Día de la Semana', fontsize=14, fontweight='bold')

# Añade los valores encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}',
            ha='center', va='bottom', fontweight='bold')

# Rotación de etiquetas del eje X
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 4: INGRESOS POR MES
# ============================================
```

```python
print("\nCreando gráfico 4: Ingresos por Mes...")

# Prepara los datos
ingresos_mes = df.groupby('month').size()

# Ordena los meses correctamente
orden_meses = ['August', 'September', 'October', 'November', 'December']
meses_español = ['Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

ingresos_mes = ingresos_mes.reindex(orden_meses)

# Crea la figura
plt.figure(figsize=(12, 6))

# Gráfico de barras con línea
bars = plt.bar(meses_español, ingresos_mes.values, color='#2ca02c', edgecolor='black', linewidth=2, alpha=0.7)
plt.plot(meses_español, ingresos_mes.values, color='red', marker='o', linewidth=2, markersize=10, label='Tendencia')

# Etiquetas y título
plt.xlabel('Mes', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.title('Tendencia de Ingresos a la Plataforma por Mes', fontsize=14, fontweight='bold')

# Añade los valores encima de cada barra
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}',
            ha='center', va='bottom', fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 5: DISTRIBUCIÓN DE ACTIVIDAD DE DOCENTES (BOXPLOT)
# ============================================
```

```python
print("\nCreando gráfico 5: Distribución de Actividad de Docentes...")

# Prepara los datos
ingresos_por_docente = df.groupby('userid').size()

# Crea la figura
plt.figure(figsize=(10, 6))

# Boxplot muestra la distribución: mínimo, cuartiles, máximo
# Stripplot añade los puntos individuales
sns.boxplot(y=ingresos_por_docente, color='lightblue', width=0.5)
sns.stripplot(y=ingresos_por_docente, color='red', alpha=0.6, size=6, jitter=True)

# Etiquetas y título
plt.ylabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.title('Distribución de Ingresos por Docente\n(Box: Q1-Q3 | Línea: Mediana | Puntos: Docentes)', 
          fontsize=12, fontweight='bold')

# Añade líneas horizontales para referencia
plt.axhline(y=ingresos_por_docente.mean(), color='green', linestyle='--', linewidth=2, label=f'Promedio: {ingresos_por_docente.mean():.0f}')
plt.axhline(y=ingresos_por_docente.median(), color='orange', linestyle='--', linewidth=2, label=f'Mediana: {ingresos_por_docente.median():.0f}')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 6: CLASIFICACIÓN DE DOCENTES POR NIVEL DE ACTIVIDAD (PASTEL)
# ============================================
```

```python
print("\nCreando gráfico 6: Clasificación de Docentes por Actividad...")

# Prepara los datos
ingresos_por_docente = df.groupby('userid').size()

muy_activos = (ingresos_por_docente >= ingresos_por_docente.quantile(0.75)).sum()
activos = ((ingresos_por_docente >= ingresos_por_docente.quantile(0.50)) & 
           (ingresos_por_docente < ingresos_por_docente.quantile(0.75))).sum()
moderados = ((ingresos_por_docente >= ingresos_por_docente.quantile(0.25)) & 
             (ingresos_por_docente < ingresos_por_docente.quantile(0.50))).sum()
poco_activos = (ingresos_por_docente < ingresos_por_docente.quantile(0.25)).sum()

# Datos para el gráfico de pastel
categorias = ['Muy Activos\n(75-100%)', 'Activos\n(50-75%)', 'Moderados\n(25-50%)', 'Poco Activos\n(0-25%)']
valores = [muy_activos, activos, moderados, poco_activos]
colores_pastel = ['#2ecc71', '#3498db', '#f39c12', '#e74c3c']

# Crea la figura
plt.figure(figsize=(10, 8))

# Gráfico de pastel (pie chart)
# autopct='%1.1f%%' muestra el porcentaje
wedges, texts, autotexts = plt.pie(valores, 
                                     labels=categorias, 
                                     autopct='%1.1f%%',
                                     colors=colores_pastel,
                                     startangle=90,
                                     textprops={'fontsize': 11, 'fontweight': 'bold'})

# Mejora el estilo del porcentaje
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

plt.title('Clasificación de Docentes por Nivel de Actividad', fontsize=14, fontweight='bold', pad=20)

# Añade leyenda con cantidad de docentes
plt.legend([f'{cat}: {val} docentes' for cat, val in zip(categorias, valores)], 
          loc='upper left', bbox_to_anchor=(0.85, 1))

plt.tight_layout()
plt.show()
```

```python
# ============================================
# GRÁFICO 7: INGRESOS POR FECHA (TENDENCIA TEMPORAL)
# ============================================
```

```python
print("\nCreando gráfico 7: Tendencia Temporal de Ingresos...")

# Prepara los datos: agrupa por fecha y cuenta
ingresos_fecha = df.groupby('date').size().reset_index(name='Ingresos')

# Crea la figura
plt.figure(figsize=(16, 6))

# Gráfico de línea
plt.plot(ingresos_fecha['date'], ingresos_fecha['Ingresos'], 
         color='steelblue', linewidth=2, marker='o', markersize=4)

# Rellena el área bajo la línea
plt.fill_between(ingresos_fecha['date'], ingresos_fecha['Ingresos'], alpha=0.3, color='steelblue')

# Etiquetas y título
plt.xlabel('Fecha', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Ingresos', fontsize=12, fontweight='bold')
plt.title('Tendencia de Ingresos a la Plataforma (Agosto - Diciembre 2024)', fontsize=14, fontweight='bold')

# Rotación de fechas en el eje X
plt.xticks(rotation=45, ha='right')

# Añade una línea de promedio
promedio = ingresos_fecha['Ingresos'].mean()
plt.axhline(y=promedio, color='red', linestyle='--', linewidth=2, label=f'Promedio: {promedio:.0f}')

plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
